```python
data_path = '../../data/calendars/2324-academic-calendar-list-view.xlsx'
data = pd.read_excel(data_path)
data['file'] = '2324-academic-calendar-list-view.txt'
```

```python
#extract fall data
fall = data.iloc[4:41]
fall.columns = ['Date', 'no1','no2','Day', 'Event','file']
fall.drop(['no1','no2'], axis=1, inplace=True)
fall['Event'] = fall['Event'].apply(lambda x: x.split(';')[0]).apply(lambda x: x.split('(')[0]).apply(lambda x: x.strip())
fall['question'] = 'For Fall 2023, When is ' + fall['Event'] + '?'
fall['answer'] = fall['Date'].apply(lambda x: x.strftime('%B %d'))  
```

```python
#extract spring data
spring24 = data.iloc[48:84]
spring24.columns = ['Date', 'no1','no2','Day', 'Event','file']
spring24.drop(['no1','no2'], axis=1, inplace=True)
spring24['Event'] = spring24['Event'].apply(lambda x: x.split(';')[0]).apply(lambda x: x.split('(')[0]).apply(lambda x: x.strip())
spring24['question'] = 'For Spring 2024, When is ' + spring24['Event'] + '?'
spring24['answer'] = spring24['Date'].apply(lambda x: x.strftime('%B %d'))  
spring24
```

```python
#extract spring data
summer24 = data.iloc[91:117]
summer24.columns = ['Date', 'no1','no2','Day', 'Event','file']
summer24.drop(['no1','no2'], axis=1, inplace=True)
summer24['Event'] = summer24['Event'].apply(lambda x: x.split(';')[0]).apply(lambda x: x.split('(')[0]).apply(lambda x: x.strip())
summer24['question'] = 'For Summer 2024, When is ' + summer24['Event'] + '?'
summer24['answer'] = summer24['Date'].apply(lambda x: x.strftime('%B %d'))  
summer24
```

```python
#extract spring data
summer224 = data.iloc[122:133]
summer224.columns = ['Date', 'no1','no2','Day', 'Event','file']
summer224.drop(['no1','no2'], axis=1, inplace=True)
summer224['Event'] = summer224['Event'].apply(lambda x: x.split(';')[0]).apply(lambda x: x.split('(')[0]).apply(lambda x: x.strip())
summer224['question'] = 'For Summer 2 2024, When is ' + summer224['Event'] + '?'
summer224['answer'] = summer224['Date'].apply(lambda x: x.strftime('%B %d'))  
summer224
```

```python
#concatenate all dataframes
all = pd.concat([fall, spring24, summer24, summer224], axis=0)
all = all[['question', 'answer', 'file']]
all['type'] = 'when'
#reorder columns
all = all[['type','question', 'answer','file']]
all.to_csv('../../data/processed_qa/2324-academic-calendar-list-view.csv', index=False)
```

```python
data_path = '../../data/commencement_schedule.csv'
data = pd.read_csv(data_path,sep='|',quotechar='"')
# data['file'] = 'commencement_schedule_data.txt'
data['location'] = data['location'].apply(lambda x: x.replace('\"',''))
data['department'] = data['department'].apply(lambda x: x.replace('schedule ',''))
#strip leading and trailing white spaces
data['location'] = data['location'].apply(lambda x: x.strip())
data['department'] = data['department'].apply(lambda x: x.strip())
data['contact'] = data['contact'].apply(lambda x: x.strip())
```

```python
#create an empty dataframe
when_qa = pd.DataFrame(columns=['type','question','answer','file'])
when_qa['question'] = 'When is the ' + data['department'] + '?'
when_qa['answer'] = data['time']
when_qa['type'] = 'when'
when_qa['file'] = 'commencement_schedule_data.txt'
when_qa
```

```python
#create an empty dataframe
where_qa = pd.DataFrame(columns=['type','question','answer','file'])
where_qa['question'] = 'Where is the ' + data['department'] + '?'
where_qa['answer'] = data['location']
where_qa['type'] = 'where'
where_qa['file'] = 'commencement_schedule_data.txt'
where_qa
```

```python
#create an empty dataframe
who_qa = pd.DataFrame(columns=['type','question','answer','file'])
who_qa['question'] = 'Who is the contact for ' + data['department'] + '?'
who_qa['answer'] = data['contact']
who_qa['type'] = 'who'
who_qa['file'] = 'commencement_schedule_data.txt'
who_qa
```

```python
#concatenate all dataframes
all = pd.concat([when_qa, where_qa, who_qa], axis=0)
all.to_csv('../../data/processed_qa/commencement_schedule_data.csv', index=False)
```
