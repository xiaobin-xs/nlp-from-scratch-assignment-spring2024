### split one file into different txt

````python
def save_rows_as_files(filename, input_folder, output_folder):
    with open(input_folder + filename, 'r') as file:
        for i, row in enumerate(file, start=1):
            # Remove leading and trailing whitespaces from the row
            row = row.strip()
            # Skip empty rows
            if not row:
                continue
            # Create a new text file for each row
            f = filename.split('.')[0]
            output_file = output_folder + f'{f}_row_{i}.txt'
            with open(output_file, 'w') as output:
                output.write(row)
````
